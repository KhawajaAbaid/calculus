```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = lambda x: 7 * x**2
```

```python
y_grad = lambda x: 14 * x
```

```python
inputs = np.linspace(-5, 5, 10)
inputs
```

```python
plt.plot(inputs, y(inputs), label="y = 7x^2")
plt.plot(inputs, y_grad(inputs), label="dy/dx = 14x")
plt.grid(True)
plt.legend()
plt.show()
```
