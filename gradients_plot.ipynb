```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = lambda x: 7 * x**2
```

```python
y_grad = lambda x: 14 * x
```

```python
inputs = np.linspace(-5, 5, 10)
inputs
```

```python
x_squared_grad_fn = lambda x: 2 * x
```

```python
plt.style.use("fivethirtyeight")
plt.plot(inputs, y(inputs), label="$ y = 7x^2 $")
plt.plot(inputs, y_grad(inputs), label="$ \delta y/ \delta x = 14x$")
# Show that the derivatis of 7x^2 are 7 times as big as that of x^2
plt.plot(inputs, np.square(inputs), label="$x^2$")
plt.plot(inputs, x_squared_grad_fn(inputs),
         label="$\delta$ $(x^2)/ \delta x = 2x$")
plt.grid(True)
plt.legend()
plt.show()
```
