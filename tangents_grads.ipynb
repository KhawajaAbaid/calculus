```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from collections import namedtuple
```

```python
# plt.style.use("fivethirtyeight")
plt.style.use("ggplot")
```

##### Curves

```python
# Curve 1
y_1 = lambda x: 1 / (2 * x) + 3

# Curve 2
y_2 = lambda x: 2 * x**2 + 2
```

```python
points = np.linspace(-5, 5, 20)
```

```python
Point = namedtuple('Point', ['x', 'y'])

# points of intersection between two curves
intersec_p_1 = Point(0, y_2(0))
intersec_p_2 = Point(-1/4, y_2(-1/4))

# equation of the tangent line
tangent_line = lambda x: (-1/2) * x + 2
```

```python
fig = plt.figure(figsize=(6, 5))
main_ax = fig.add_axes([0, 0, 1, 1])
main_ax.plot(points, y_1(points), label="$ y = 1/2x + 3 $")
main_ax.plot(points, y_2(points), label="$ y = 2x^2 + 2 $")
main_ax.plot(points, tangent_line(points), label="tangent", color="purple")

maginfy_tangent = fig.add_axes([0.3, 0.3, 0.2, 0.2])
maginfy_tangent.plot(points[5:9], y_1(points[5:9]))
maginfy_tangent.plot(points[5:9], tangent_line(points[5:9]), color="purple")
maginfy_tangent.set_title("tangent")

maginfy_tangent_2 = fig.add_axes([0.6, 0.5, 0.2, 0.2])
maginfy_tangent_2.plot(points[8:11], y_2(points[8:11]), color="cyan")
maginfy_tangent_2.plot(points[8:11], tangent_line(points[8:11]), color="purple")
maginfy_tangent_2.set_title("tangent 2")

main_ax.grid(True)
main_ax.legend()
plt.show()
```
